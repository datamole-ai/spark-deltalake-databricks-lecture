```python
TABLE_NAME = "data/01-table"
```

```python
!rm -rf {TABLE_NAME}
```

```python
import pandas as pd
from deltalake.writer import write_deltalake

df = pd.DataFrame({"x": [1, 2, 3]})

write_deltalake(TABLE_NAME, df)
```

```python
df2 = pd.DataFrame({"x": [9, 8, 10]})

write_deltalake(TABLE_NAME, df2, mode="append")
```

```python
df3 = pd.DataFrame({"x": [55, 66, 77]})

write_deltalake(TABLE_NAME, df3, mode="overwrite")
```
